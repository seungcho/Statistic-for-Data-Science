## **1.1 정형화된 데이터 요소**

> **실제 업무에선 마치 DB에서 뽑은 듯한 원시 데이터들을 가공하여 정형화된 형태로 변환하거나 혹은 애초에 특정한 연구 목적을 가지고 수집해야한다.**

### **수치형 데이터와 범주형 데이터, 정형 데이터의 가장 기본이 되는 종류이다.**

**수치 데이터는 풍속이나 지속 시간 같은 연속형 데이터, 그리고 사건의 발생 빈도 같은 이산 데이터가 있다.**

> **순서형 데이터는 범주형 데이터 중 평점(1,2,3,4,5)이 순서형 데이터의 대표적인 예라고 할 수 있다.**

## **1.3 위치 추정**

**데이터를 살펴보는 가장 기초적인 단계는 각 피처(=변수)의 대푯값을 구하는 것이다.**

> **이는 곧 대부분의 값이 어디쯤에 위치하는지(중심경향성)을 나타내는 추정값이다.**

**데이터를 요약하려면 그냥 데이터의 평균을 구하기만 하면 되지 않느냐고 생각할지도 모른다. 하지만 평균이 데이터의 중간을 대표하는 가장 좋은 방법은 아니다.**

>  **통계학이라는 분야는 궁극적으로 불확실성을 이해하고자 하는 의미에서 추정값 이라는 용어를 사용하는 반면, 데이터 과학은 구체적인 비즈니스나 조직의 목표치에 관심을 두기 때문에 측정 지표라 부른다.**

> **통계학에서의 추정은 어떤 참값을 어림잡아 추측한다는 의미가 강한 반면, 데이터 과학에서의 측정 지표는 어떠한 기준값(혹은 목표값)과의 거리(혹은 오차)를 표현하는 계측량이라 할 수 있다.**

### **1.3.1 평균**

**평균은 가장 기본적인 위치 추정 방법이다.**

> **모든 값의 총합을 값의 개수로 나눈 값**

![test](./img/평균.jpg)

**절사평균은 값들을 크기 순으로 정렬한 후, 양끝에서 일정 개수의 값들을 삭제한 뒤 남은 값들을 가지고 구한 평균을 말한다.**

> **극단값의 영향을 제거하는 장점이 있다.**

![test](./img/절사평균.jpg)

**가중평균은 확률을 수반하는 평균 산출 시 활용한다.**

> **어떤 값들이 본래 다른 값들에 비해 큰 변화량을 갖을 때, 이러한 관측값에 대해 더 작은 가중치를 줄 수 있다.**

> **데이터를 수집할 때, 우리가 관심 있는 서로 다른 대조군에 대해서 항상 똑같은 수가 얻어지지는 않는다. 모든 사용자 그룹에 대해 정확히 같은 비율을 반영하는 데이터를 수집하기는 매우 어렵다. 이를 보정하기 위해서, 데이터가 부족한 소수 그룹에 대해 더 높은 가중치를 적용할 필요도 있을 것이다.**

![test](./img/가중평균.jpg)

![test](./img/가중평균2.png)

### **1.3.2 중간값과 로버스트 추정**

**데이터를 일렬로 정렬했을 때, 한가운데 위치하는 값을 중간값이라 한다.**

**데이터에 매우 민감한 평균보다는 중간값이 많이 분포되어 있는 경우, 위치추정에 유리하다.**

> **중간값은 결과를 왜곡할 수도 있는 특잇값(극단값)들의 영향을 받지 않으므로 로버스트한 위치 추정 방법이라 알려져 있다.**

### **특잇값**

**특잇값은 어떤 데이터 집합에서 다른 값들과 매우 멀리 떨어져 있는 값들을 말한다.**

> **1. 데이터 값 자체가 유효하지 않는 경우**

> **2. 데이터 값이 잘못된 경우**

> **3. 정말 특이한 경우의 값**

**중간값 뿐만 아니라 절사평균 역시 특잇값의 영향을 줄이기 위해 많이 사용한다. 하지만 절사평균의 경우 데이터가 적을땐 효과적이지 않다.**

### **1.3.3 예제: 인구에 따른 살인 비율의 위치 추정**

**미국 각 주의 인구와 살인 비율(인구 10만 명당 연간 살인 사건)을 보여준다.**

```python
# python을 이용해 인구의 평균, 절사평균, 중간값을 계산해본다.
import pandas as pd 
from scipy import stats # 절사평균을 구하기 위한 사이파이의 stats 패키지 불러오기
import numpy as np
```

## **1.4 변이 추정**

**변이는 데이터 값이 얼마나 밀집해 있는지 혹은 퍼져 있는지를 나타내는 산포도를 나타낸다.**

> **변이를 측정하고, 이를 줄이고, 실제 변이와 랜덤을 구분하고, 실제 변이의 다양한 요인들을 알아보고, 변이가 있는 상황에서 결정을 내리는 등, 통계의 핵심에는 변이가 있다.**

### **1.4.1 표준편차와 관련 추정값들**

**가장 대표적으로 사용되는 변위 추정들은 관측 데이터와 위치 추정값(평균,중앙값 등등..)사이의 차이, 즉 편차를 기본으로 한다.**

> **이러한 편차는 데이터가 위치 추정값 주변으로 얼마나 퍼져 있는지를 말해준다.**

**이러한 변이를 측정하는 한 가지 방법은 편차들의 대푯값을 추정하는 것이다. 이를 평균절대편차라 한다.**

![test](./img/평균절대편차.jpg)

**가장 유명한 변이 추정 방법은 제곱 편차를 이용하는, 분산과 표준편차이다.**

> **표준편차는 원래 데이터와 같은 척도(scale)에 있기 때문에 분산보다 훨씬 해석하기 쉽다.**

> **제곱한 값이 절대값보다 통계 모델을 다루는 데 더 편리하다는 이론이 있음으로 분산과 표준편차를 더 많이 사용한다.**

> **분산,표준편차,평균절대편차,중위절대편차 모두 동일한 추정은 아니지만, 모두 데이터가 정규분포에서 왔다고 가정하고 있다.**

![test](./img/분산.png)

### **자유도 n 아니면 n-1?**

**만약 분산 수식에 n을 분모로 사용한다면, 모집단의 분산과 표준편차의 참값을 과소평가하게 된다. 이를 편향 추정이라 부른다. 하지만 만약 n 대신 n-1로 나눈다면, 분산은 비편향 추정이 된다.**

> **표본의 분산은 모집단의 분산을 과소평가하여 (표본의 분산 < 모집단의 분산)와 같은 상태가 되기 때문에 이를 보정해주기 위해서 표본분산의 분모를 작게 만들어 전체 표본분산을 크게 만들었다.**

> **n을 사용하는 게 편향 추정이 되는지에 대한 이유는 추정값을 계산할 때 제약 조건의 개수를 의미하는 자유도에 대해 언급할 필요가 있다.**

> **이 경우, 표준편차는 표본 평균에 따른다는 하나의 제약 조건을 가지고 있기 때문에 n-1의 자유도를 갖는다. 보통 대부분의 경우 데이터 과학자들은 자유도에 크게 신경쓰지 않아도 된다.**

![test](./img/표본분산.png)

### **1.4.2 백분위수에 기초한 추정**

**변위를 추정하는 또 다른 접근은 정렬된 데이터가 얼마나 퍼져 있는지를 보는 것이다. 정렬(순위) 데이터를 나타내는 통계량을 순서통계량이라 부른다.**

> **가장 기본이 되는 측도는 가장 큰 값과 작은 값의 차이를 나타내는 범위이다. (하지만 특잇값에 매우 민감하고 그리 유용하지 않다.)**

**변위를 특정하는 가장 대표적인 방법은 사분위수범위(IQR)라는 25번째 백분위수와 75번째 백분위수의 차이를 보는 것이다.**

![test](./img/백분위.png)

### **1.4.3 예제: 주별 인구의 변위 추정**

**python을 통해 표준편차, 사분위범위(IQR), 중위절대편차(MAD)와 같은 주별 인구의 변위 추정값들을 계산해본다.**

```python
state['Population'].var() # 인구 칼럼의 분산
```

```python
state['Population'].std() # 인구 칼럼의 표준편차
```

```python
state['Population'].mad() # 평균 절대 편차
```

```python
# 표준편차는 MAD의 거의 두 배가 된다. 표준편차는 특잇값에 민감하므로 이 결과는 놀라운 것이 아니다.
def get_median_absolute_deviation(data): # 중위절대편차를 구하기 위한 함수
    return np.median(np.absolute(data - np.median(data, axis=0)), axis=0)
get_median_absolute_deviation(state['Population'])
```

```python
from scipy.stats import iqr # 사분위수를 구하기 위한 패키지 임포트
iqr(state['Population']) # 25% 되는 위치와 75% 되는 위치의 사이를 말한다.
```

```python
# 누적도수 구하기
# 누적도수는 자기 앞 도수를 다 더해서 도수를 만들면 된다.
accumul_weight = []
previous = 0
for i in hist:
    previous += i
    accumul_weight.append(previous)
print('누적도수:',accumul_weight)
```

## **1.7 상관관계**

**많은 경우 모델링 프로젝트에서 탐색적 데이터 분석이라 하면 예측값들 간의 혹은 예측값과 목푯값과의 상관관계를 조사해보는 것을 빼놓을 수 없다.**

**상관계수란 수치적 변수들 간에 어떤 관계가 있는지를 나타내기 위해 사용되는 측정량이다.(-1 ~ +1)**

> **두 변수 사이의 상관관계를 항상 같은 척도에 놓고 추정하는 것이다.**

> **상관계수는 항상 +1(완전 양의 상관관계), -1(완전 음의 상관관계), 0(선형관계가 없음)의 사이의 값으로 존재**

> **변수들이 선형적인 관계를 갖지 않을 경우, 상관계수는 더 이상 유용한 측정 지표가 아니다.**

![test](./img/상관계수.png)

**상관관계를 추정하는 다른 방법: 오래전부터 통계학자들은 스피어먼의 로(rho)나 켄들의 타우(kendall's tau)와 같은 다른 형태의 상관계수들을 만들어냈다. 이들은 데이터의 순위를 기초로 하는 상관계수이다.**

> **값 자체보다 값의 순위를 이용하기 때문에 이러한 추정법은 특잇값에 좀 더 로버스트하며 비선형 관계도 다룰 수 있다.**

**하지만 데이터 과학자들은 일반적으로 피어슨 상관계수(일반적인 상관계수), 혹은 이것의 로버스트한 다른 버전들을 탐색 분석에 주로 사용한다.**

> **순위 기반 추정법은 보통 데이터의 크기가 작고 어떤 특별한 가설검정이 필요할 때 주로 사용된다.**

### **1.8.1 범주형 변수 대 범주형 변수**

**분할표는 두 범주형 변수를 요약하는 데 효과적인 방법이다. 즉, 범주별 빈도수를 기록한 표를 뜻한다.**

> **이 표는 빈도와 백분율을 보여주며 분할표를 통해 각 열의 빈도나 전체 백분율을 알 수 있다.**

**파이선에서 빈도표를 만드는 방법은 여러가지가 있지만, 그 중 하나가 pandas의 crosstab 함수를 이용한다.**

> **아래처럼 dataframe의 칼럼(Series)을 파라미터로 넘겨주면 빈도표가 만들어진다.**

```python
# 대출 등급과 상황에 대한 데이터셋 불러오기
loans = pd.read_csv('./data/lc_loans.csv')
```

```python
loans.head()
```

```python
# dataframe의 칼럼(Series)을 파라미터로 넘겨주면 빈도표가 만들어진다.
# 아래 코드는 status를 기준으로 하여 grade의 빈도를 구한다.
cross_tab = pd.crosstab(loans['status'],loans['grade'],margins=True)
cross_tab
```

```python
# 만들어진 crosstab에 접근하는 방법
# crosstab을 통해 생성한 객체는 dataframe 타입이다. 
# 따라서 인덱스의 값으로 해당 인덱스의 원소값을 얻는 loc 함수를 통해 crosstab 원소에 접근할 수 있다.
cross_tab.loc[['Late'],['B']] # .loc[[행],[열]]
```
